```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import json
```

```python
with open('OriginalsAnsweredDraw.json') as file:
    result = json.load(file)
```

```python
result
```

```python
data = [result['views'][0]['originals'],result['views'][0]['no-answered']]
```

```python
data
```

```python
annotations = [round(result['views'][0]['percentage-answered'],2),round(result['views'][0]['percentage-isolated'],2)]
```

```python
annotations
```

```python
preparado = pd.Series(data[0])
```

```python
preparado2 = pd.Series(data[1])
```

```python
fig, ax = plt.subplots()
```

```python
ax=preparado.plot.barh(label="Sin respuesta " +str(annotations[0])+'%', color='darkorange')
```

```python
ax=preparado2.plot.barh(alpha = 1,label="Con respuesta " + str(annotations[1])+'%',color='darkblue')
```

```python
ax.legend()
```

```python
plt.title("Originales respondidos y originales no respondidos")
```
