```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import json
```

```python
with open('resultados_definitivos/ConnectedComponents.json') as file:
    output = json.load(file)
```

```python
values=output['views'][0]['top5']
```

```python
values
```

```python
size=[]; freq=[]
```

```python
for line in values:
    size.append(line['size'])
    freq.append(line['freq'])
```

```python
x=size[::-1]; y=freq[::-1]
```

```python
cdf=np.cumsum(y)
```

```python
percentage=[]
for line in cdf:
    percentage.append(line/cdf[len(cdf)-1])
```

```python
plt.plot(x,percentage)
```

```python
plt.xscale('log')
```

```python
plt.xlabel('TamaÃ±o de isla')
```

```python
plt.ylabel('Frecuencia')
```

```python
plt.title('DistribuciÃ³n acumulada del tamaÃ±o de las islas')
```
