```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import json
```

```python
with open('resultados_definitivos/DegreeOriginalsTime/DegreeOriginalsWeekHour.json') as file:
    output=json.load(file)
```

```python
result=output['views']
result
```

```python
times=[];data=[]
```

```python
i=0
for line in result:
    #a=(19799917+(i*60*60*24*7))
    #print(a)
    #if(a == line['time']):
        b=[]
        b.append(line['originals'])
        b.append(line['answers'])
        times.append(line['time']/60/60/24/7)
        b.append(line['original-per-answer'])
        data.append(b)
        i=i+1
```

```python
df = pd.DataFrame(data,columns=['Originales','Respuestas','Relacin'],index=times)
```

```python
print(data)
```

```python
df.plot(color=['darkblue','darkorange','green'])
```

```python
plt.yscale('log')
```
