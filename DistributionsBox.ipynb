```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import json
```

```python
with open('DistributionsDraw.json') as file:
    output = json.load(file)
```

```python
withOriginals=output['views'][0]['degDist']
```

```python
answers=output['views'][0]['answerDistribution']
```

```python
oriExpanded = []; ansExpanded = []
```

```python
withOriginals
```

```python
for line in withOriginals:
    for i in range(line['freq']):
        oriExpanded.append(line['degree'])
```

```python
for line in answers:
    for i in range(line['freq']):
        ansExpanded.append(line['weight'])
```

```python
plt.boxplot([oriExpanded,ansExpanded])
```
