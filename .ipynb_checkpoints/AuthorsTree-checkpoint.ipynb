```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import json
```

```python
with open('resultados_definitivos/AuthorsTree.json') as file:
    output = json.load(file)
```

```python
threads=output['views'][0]['total']
```

```python
threads
```

```python
sizes=[]; authors=[]
```

```python
for line in threads:
    sizes.append(line['size'])
    authors.append(line['authors'])
```

```python
plt.figure()
```

```python
total = sns.regplot(x=sizes ,y=authors)
```

```python
plt.xlabel('TamaÃ±o')
```

```python
plt.ylabel('NÃºmero de autores')
```

```python
plt.title('DistribuciÃ³n del tamaÃ±o y los autores por cada isla')
```

```python
plt.figure()
first = sns.regplot(x=sizes ,y=authors, x_estimator=np.mean)
plt.xlim([0,50])
plt.ylim([0,50])
plt.xlabel('TamaÃ±o')
plt.ylabel('NÃºmero de autores')
plt.title('Del cero al cincuenta')
```
