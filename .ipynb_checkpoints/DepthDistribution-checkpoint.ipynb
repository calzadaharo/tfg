```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import json
```

```python
with open('resultados_definitivos/DepthDistribution.json') as file:
    output = json.load(file)
```

```python
threads=output['views'][0]['threads']
```

```python
threads
```

```python
levels = []; freqs = []
```

```python
for line in threads:
    if(line['thread'] != -1):
        levels.append(line['thread']+1)        
```

```python
for line in threads:
    if(line['thread'] != -1):
        freqs.append(line['freq'])  
```

```python
levels
```

```python
freqs
```

```python
cdf = np.cumsum(freqs)
```

```python
cdf
```

```python
proportional=(cdf-cdf[0])/(cdf[len(cdf)-1]-cdf[0])
```

```python
proportional
```

```python
plt.plot(levels,proportional,color='darkblue')
```

```python
plt.title('DistribuciÃ³n acumulada del nivel de las respuestas')
```

```python
plt.xlabel('Niveles')
```

```python
plt.ylabel('Frecuencia')
```

```python
plt.xscale('log')
plt.ylim([0,1])
```
