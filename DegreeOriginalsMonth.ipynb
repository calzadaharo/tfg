```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import json
```

```python
with open('resultados_definitivos/DegreeOriginalsTime/DegreeOriginalsMonthHour.json') as file:
    output=json.load(file)
```

```python
result=output['views']
result
```

```python
times=[];data=[]
```

```python
i=0
for line in result:
    #a=(19799917+(i*60*60*24*30))
    #print(a)
    #if(i < line['time']):
        b=[]
        b.append(line['originals'])
        b.append(line['answers'])
        times.append(line['time']/60/60/24/30)
        b.append(line['original-per-answer'])
        data.append(b)
        #i=line['time']
```

```python
df = pd.DataFrame(data,columns=['Originales','Respuestas','Relación'],index=times)
```

```python
df.plot(color=['darkblue','darkorange','green'])
```
