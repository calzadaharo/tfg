```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import json
```

```python
with open('resultados_definitivos/DegreeOriginalsTime/DegreeOriginalsDayHour.json') as file:
    output=json.load(file)
```

```python
output
```

```python
originals=[]; answers=[]; times=[]; degrees=[];data=[]
```

```python
for result in output['views']:
    line=[]
    line.append(result['originals'])
    line.append(result['answers'])
    times.append(result['time']/60/60/24)
    line.append(result['original-per-answer'])
    data.append(line)
```

```python
df = pd.DataFrame(data,columns=['Originales','Respuestas','Relación'],index=times)
```

```python
df.plot(color=['darkblue','darkorange','green'])
```

```python
plt.yscale('log')
```

```python
plt.title('Relación entre mensajes originales y respuestas cada día')
```
