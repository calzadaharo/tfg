```python
with open('resultados_definitivos/DegreeOriginals.json') as file:
    data = json.load(file)
```

```python
for line in data['views']: print(line)
```

```python
nombres=['Originales', 'Respuestas']
```

```python
datos = [data['views'][0]['originals'],data['views'][0]['answers']]
```

```python
preparado=pd.Series(datos,index=nombres)
```

```python
ax=preparado.plot.bar(rot=0,title='Originales vs respuestas',color=['darkblue','darkorange'])
```

```python
anotaciones = [round(data['views'][0]['percentage-originals'],2),round(data['views'][0]['percentage-answers'],2)]
```

```python
anotaciones
```

```python
ax.annotate(str(anotaciones[0])+'%', xy=(ax.patches[0].get_x() + ax.patches[0].get_width() / 2,ax.patches[0].get_height()),
           va='center',ha='center', xytext=(0, 6),textcoords='offset points')
```

```python
ax.annotate(str(anotaciones[1])+'%', xy=(ax.patches[1].get_x() + ax.patches[1].get_width() / 2,ax.patches[1].get_height()),
           va='center',ha='center', xytext=(0, 6),textcoords='offset points')
```
