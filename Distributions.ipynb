```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import json
```

```python
with open('DistributionsDraw.json') as file:
    output = json.load(file)
```

```python
withOriginals=sorted(output['views'][0]['degDist'], key=lambda x: x['degree'])
```

```python
withOriginals
```

```python
output['views'][0]['answerDistribution']
```

```python
outputSorted = sorted(output['views'][0]['answerDistribution'], key=lambda x: x['weight'])
```

```python
outputSorted
```

```python
degrees=[]; freqs=[]
```

```python
for line in outputSorted:
    degrees.append(line['weight'])
    freqs.append(line['freq'])
```

```python
newDegrees=[]; newFreqs=[]
```

```python
for original in withOriginals:
    newDegrees.append(original['degree'])
    newLine=[]
    newLine.append(original['freq'])
    for answered in outputSorted:
        if(answered['weight']==original['degree']):
            newLine.append(answered['freq'])
            break
    if(len(newLine) == 1):
        newLine.append(0)
    newFreqs.append(newLine)
```

```python
print(newFreqs)
```

```python
print(newDegrees)
```

```python
data = pd.DataFrame(newFreqs, index=newDegrees,columns=['Todos', 'Sólo las respuestas']) #dataFrame que incluya todos los originales
```
