```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import json
```

```python
with open('resultados_definitivos/ConnectedComponents.json') as file:
    output = json.load(file)
```

```python
output 618765
```

```python
values=output['views'][0]['top5']
newValues=values[::-1]
print(newValues)
```

```python
freqs = []
```

```python
for line in newValues:
    for i in range(line['freq']):
        freqs.append(line['size'])
print(freqs)
```

```python
df = pd.DataFrame(freqs)
```

```python
data = plt.boxplot(freqs, boxprops=dict(color='darkblue'),
            capprops=dict(color='darkblue'),
            whiskerprops=dict(color='darkblue'),
            flierprops=dict(color='darkblue', markeredgecolor='darkblue'),
            medianprops=dict(color='darkblue'))
```

```python
plt.yscale('log')
```

```python
plt.title('Distribución del número de mensajes en cada conversación')
```
